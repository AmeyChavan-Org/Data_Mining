Q1.Draw a line plot for the population of India.

```python
year=df["year"]
India=df["India"]
plt.plot(year,India)
plt.title('The population of India.')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

Q2. Draw a line chart for USA, India, China, Indonesia, and Brazil population in the duration of 1955 to 2020.

```python
United=df["United States"]
China=df["China"]
Indonesia=df["Indonesia"]
Brazil=df["Brazil"]
india=df["India"]

plt.plot(year,China, color='red', marker='o')
plt.plot(year,Brazil, color='black', marker='>')
plt.plot(year,Indonesia, color='orange', marker='<')
plt.plot(year,Brazil, color='blue', marker='^')
plt.plot(year,india, color='green', marker='o')
plt.plot(year,United, color='pink', marker='o')
plt.title('population 1955-2020')
plt.xlabel('year')
plt.ylabel('population ')
plt.show()
```

Q4) Draw a group bar plot for USA, India, China, Indonesia, and Brazil population in year
[1980, 1990, 2000, 2010, 2010]

Q7.Draw a pie chart to represent the population in 2020 for the following countries: USA,India, China, Indonesia, and Brazil

```python
countries = ['USA', 'India', 'China', 'Indonesia', 'Brazil']
population = [331002651, 1380004385, 1444216107, 273523615, 212559417]  # Population in 2020
# Plotting
plt.pie(population, labels=countries, autopct='%1.1f%%',wedgeprops={"edgecolor":"black"},radius=1.5)
plt.title('Population in 2020')
plt.legend(loc = 'upper right')
plt.show()
```
